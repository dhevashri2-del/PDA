```python
 #accessing array index
import numpy as np
arr=np.array([[-1,2,0,4],
             [4,-0.5,6,0],
             [2.6,0,7,8],
             [3,-7,4,2.0]])
arr2=arr[:2,::2]
print("first 2 rows and alternate columns(0 and 2):\n", arr2)
```

```python
# BASIC ARRAY OPERATIONS
import numpy as np
 
# defining Array 1
a = np.array([[1, 2],
              [3, 4]])
 
# defining Array 2
b = np.array([[4, 3],
              [2, 1]])
               
# adding 1 to every element
print ("Adding 1 to every element:", a + 1)
 
# subtracting 2 from each element
print ("\nSubtracting 2 from each element:", b - 2)
 
# sum of array elements
# Performing Unary operations
print ("\nSum of all array elements: ", a.sum())
 
# Adding two arrays
# Performing Binary operations
print ("\nArray sum:\n", a + b)
```

```python
# MATH OPERATIONS ON DATA TYPE ARRAY
import numpy as np
 
# First Array
arr1 = np.array([[4, 7], [2, 6]], 
                 dtype = np.float64)
                  
# Second Array
arr2 = np.array([[3, 6], [2, 8]], 
                 dtype = np.float64) 
 
# Addition of two Arrays
Sum = np.add(arr1, arr2)
print(Sum)
 
# Addition of all Array elements
Sum1 = np.sum(arr1)
print(Sum1)
 
# Square root of Array
Sqrt = np.sqrt(arr1)
print(Sqrt)
 
# Transpose of Array
Trans_arr = arr1.T
print(Trans_arr)
```

```python
# BASIC SLICING AND INDEXING
import numpy as np  
a = np.arange(20)

print("a[15] =", a[15])
print("a[-8:17:1] =", a[-8:17:1])
print("a[10:] =", a[10:])
```

```python
# MATRIX OPERATIONS IN NUMPY
import numpy as np

x = np.array([[1, 2], [4, 5]])
y = np.array([[7, 8], [9, 10]])

print("Addition:\n", np.add(x, y))
print("Subtraction:\n", np.subtract(x, y))
print("Division:\n", np.divide(x, y))
```

```python
import numpy as np

x = np.array([[1, 2], [4, 5]])
y = np.array([[7, 8], [9, 10]])

print("Element-wise multiplication:\n", np.multiply(x, y))
print("Matrix multiplication:\n", np.dot(x, y))
```

```python
import numpy as np

x = np.array([[1, 2], [4, 5]])
y = np.array([[7, 8], [9, 10]])

print("Square root:\n", np.sqrt(x))
print("Sum of all elements:", np.sum(y))

print("Column-wise sum:", np.sum(y, axis=0))
print("Row-wise sum:", np.sum(y, axis=1))
print("Transpose:\n", x.T)
```

```python
import numpy as np
# Creating NumPy arrays
# 1D array
a1 = np.array([1, 2, 3, 4, 5, 6, 7, 8])

# 2D array
a2 = np.array([[1, 2],
               [3, 4],
               [5, 6],
               [7, 8]])

# 3D array
a3 = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])

print(a1)
print(a2)
print(a3)
```

```python
# Using NumPy functions
c1 = np.zeros((3, 3))
c2 = np.ones((5, 5))
c3 = np.arange(0, 10, 2)
print(c1)
print(c2)
print(c3)

# Array Indexing
print(a1[2])      
print(a1[-1])    
print(a2[1][0]) 
print(a2[1][1])   
print(a3[0][0][0]) 

# Array Slicing
print(a1[1:4])    
print(a2[:, 1])
```

```python
#checking data types
print(type(a1))
# Parameters of array
# Shape
print(a1.shape)
print(a2.shape)
print(a3.shape)

# Rank (number of dimensions)
print(a1.ndim)
print(a2.ndim)
print(a3.ndim)
```
